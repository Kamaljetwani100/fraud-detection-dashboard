# Task_2: Preprocessing, Imbalance Handling & Feature Engineering.

### Task_2.1: Drop Columns with More Than 50% Missing Values.

```python
# Create a working copy of the dataset
df = data.copy()

# Drop columns identified in Task 1 with more than 50% missing values
df.drop(columns=columns_to_drop, inplace=True)

# Display updated shape
print("Original dataset shape:", data.shape)
print("Dataset shape after dropping high-missing columns:", df.shape)
print("Number of columns dropped:", len(columns_to_drop))
```

### Task_2.2: Impute Remaining Values Using Median (Numerical) and Mode (Categorical).

```python
# Separate numerical and categorical columns
numerical_cols = df.select_dtypes(include=["number"]).columns.tolist()
categorical_cols = df.select_dtypes(include=["object", "category"]).columns.tolist()

# Remove target column from numerical features list
if "isFraud" in numerical_cols:
    numerical_cols.remove("isFraud")

# Impute numerical columns with median
for col in numerical_cols:
    df[col].fillna(df[col].median(), inplace=True)

# Impute categorical columns with mode
for col in categorical_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)

# Verify that no missing values remain
remaining_missing = df.isnull().sum().sum()

print("Remaining missing values:", remaining_missing)
print("Numerical columns imputed:", len(numerical_cols))
print("Categorical columns imputed:", len(categorical_cols))
```

### Task_2.3: Label-Encode High-Cardinality Categorical Columns.

#### Encoding Strategy Justification

The dataset contains several categorical features such as product codes, email domains, device information, and browser identifiers. Many of these columns have a large number of unique values (high cardinality).

One-hot encoding would significantly increase the number of columns and memory usage, making the model slower and less efficient. To avoid this, Label Encoding is used to convert each category into a unique integer while preserving a compact feature space.

Tree-based algorithms such as XGBoost, LightGBM, and CatBoost can work effectively with integer-encoded categorical values, making label encoding an appropriate and scalable choice for this large fraud detection dataset.

```python
# Store label encoders for future use
label_encoders = {}

# Encode all categorical columns
for col in categorical_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col].astype(str))
    label_encoders[col] = le

print("Total categorical columns encoded:", len(categorical_cols))
print("Sample encoded columns:", categorical_cols[:10])
```

### Task_2.4: Create at Least 3 Engineered Features.

```python
# 1. Amount-to-Mean Ratio
global_mean_amt = df["TransactionAmt"].mean()
df["AmtToMeanRatio"] = df["TransactionAmt"] / global_mean_amt

# 2. Hour of Day (derived from TransactionDT, which is measured in seconds)
df["HourOfDay"] = (df["TransactionDT"] // 3600) % 24

# 3. Device Risk Flag
# If device-related columns are missing after column dropping,
# create a default flag of 0.
if "DeviceType" in df.columns and "DeviceInfo" in df.columns:
    # DeviceType and DeviceInfo were already label encoded.
    # A simple heuristic: if either field is present (non-null before encoding),
    # assign risk flag = 1.
    df["DeviceRisk"] = (
        (df["DeviceType"] >= 0) &
        (df["DeviceInfo"] >= 0)
    ).astype(int)
else:
    df["DeviceRisk"] = 0

# Display newly created features
engineered_features = [
    "AmtToMeanRatio",
    "HourOfDay",
    "DeviceRisk"
]

print("Engineered Features Created:")
for feature in engineered_features:
    print("-", feature)

display(df[engineered_features].head())
```

### Task_2.5: Save the Final Cleaned Dataset to `processed_data.csv`.

```python
# Save the fully processed dataset
output_path = "data/processed_data.csv"
df.to_csv(output_path, index=False)

# Confirm save
print(f"Processed dataset saved successfully to: {output_path}")
print("Final dataset shape:", df.shape)
print("Total engineered features added:", 3)
```

### Task_2.6: Perform Stratified 80/20 Train-Test Split.

```python
from sklearn.model_selection import train_test_split

# Separate features and target
X = df.drop("isFraud", axis=1)
y = df["isFraud"]

# Stratified train-test split (80/20)
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print("Training set shape:", X_train.shape)
print("Testing set shape:", X_test.shape)

print("\nTraining class distribution:")
print(y_train.value_counts(normalize=True) * 100)

print("\nTesting class distribution:")
print(y_test.value_counts(normalize=True) * 100)
```

### Task_2.7: Scale Numerical Features Using RobustScaler.

```python
from sklearn.preprocessing import RobustScaler

# Identify feature columns (exclude target, already removed from X)
numeric_features = X_train.select_dtypes(include=["number"]).columns.tolist()

# Initialize scaler
scaler = RobustScaler()

# Fit only on training data and transform both sets
X_train[numeric_features] = scaler.fit_transform(
    X_train[numeric_features]
)

X_test[numeric_features] = scaler.transform(
    X_test[numeric_features]
)

print("Numerical features scaled:", len(numeric_features))
print("Scaling method used: RobustScaler")
```

### Task_2.8: Apply SMOTE Only on the Training Set.

```python
from imblearn.over_sampling import SMOTE

# Final safety check: remove any remaining missing values
# Numerical columns -> fill with median
for col in X_train.columns:
    if X_train[col].isnull().sum() > 0:
        if X_train[col].dtype in ["int64", "float64"]:
            median_value = X_train[col].median()
            X_train[col] = X_train[col].fillna(median_value)
            X_test[col] = X_test[col].fillna(median_value)
        else:
            mode_value = X_train[col].mode()[0]
            X_train[col] = X_train[col].fillna(mode_value)
            X_test[col] = X_test[col].fillna(mode_value)

# Verify missing values are removed
print("Remaining missing values in X_train:", X_train.isnull().sum().sum())
print("Remaining missing values in X_test:", X_test.isnull().sum().sum())

# Class distribution before SMOTE
print("\nClass Distribution Before SMOTE:")
print(y_train.value_counts())

# Initialize SMOTE
smote = SMOTE(random_state=42)

# Apply SMOTE only to the training set
X_train_smote, y_train_smote = smote.fit_resample(
    X_train,
    y_train
)

# Class distribution after SMOTE
print("\nClass Distribution After SMOTE:")
print(y_train_smote.value_counts())

print("\nTraining shape before SMOTE:", X_train.shape)
print("Training shape after SMOTE:", X_train_smote.shape)
```

### Task_2.9: Report Class Ratio Before and After SMOTE.

```python
# Percentage distribution before SMOTE
before_smote_pct = y_train.value_counts(normalize=True) * 100

# Percentage distribution after SMOTE
after_smote_pct = y_train_smote.value_counts(normalize=True) * 100

print("=" * 60)
print("CLASS RATIO COMPARISON")
print("=" * 60)

print("\nBefore SMOTE (%):")
print(before_smote_pct)

print("\nAfter SMOTE (%):")
print(after_smote_pct)

# Create comparison table
class_ratio_comparison = pd.DataFrame({
    "Before_SMOTE (%)": before_smote_pct,
    "After_SMOTE (%)": after_smote_pct
}).round(2)

print("\nComparison Table:")
display(class_ratio_comparison)
```

# Task_3: Model Training, Comparison & Threshold Optimization.

### Task_3.1: Train LightGBM Classifier.

```python
from lightgbm import LGBMClassifier
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    average_precision_score
)

# Initialize LightGBM model
lgbm_model = LGBMClassifier(
    n_estimators=200,
    learning_rate=0.05,
    num_leaves=31,
    random_state=42,
    n_jobs=-1
)

# Train on SMOTE-balanced training data
lgbm_model.fit(X_train_smote, y_train_smote)

# Predict on test set
lgbm_pred = lgbm_model.predict(X_test)
lgbm_proba = lgbm_model.predict_proba(X_test)[:, 1]

# Store metrics
results = {}

results["LightGBM"] = {
    "Accuracy": accuracy_score(y_test, lgbm_pred),
    "Precision": precision_score(y_test, lgbm_pred),
    "Recall": recall_score(y_test, lgbm_pred),
    "F1-Score": f1_score(y_test, lgbm_pred),
    "ROC-AUC": roc_auc_score(y_test, lgbm_proba),
    "PR-AUC": average_precision_score(y_test, lgbm_proba)
}

print("LightGBM training completed successfully.")
print(results["LightGBM"])
```

### Task_3.2: Train XGBoost Classifier.

```python
from xgboost import XGBClassifier

# Initialize XGBoost model
xgb_model = XGBClassifier(
    n_estimators=200,
    learning_rate=0.05,
    max_depth=6,
    subsample=0.8,
    colsample_bytree=0.8,
    eval_metric="logloss",
    random_state=42,
    n_jobs=-1
)

# Train on SMOTE-balanced training data
xgb_model.fit(X_train_smote, y_train_smote)

# Predict on test set
xgb_pred = xgb_model.predict(X_test)
xgb_proba = xgb_model.predict_proba(X_test)[:, 1]

# Store metrics
results["XGBoost"] = {
    "Accuracy": accuracy_score(y_test, xgb_pred),
    "Precision": precision_score(y_test, xgb_pred),
    "Recall": recall_score(y_test, xgb_pred),
    "F1-Score": f1_score(y_test, xgb_pred),
    "ROC-AUC": roc_auc_score(y_test, xgb_proba),
    "PR-AUC": average_precision_score(y_test, xgb_proba)
}

print("XGBoost training completed successfully.")
print(results["XGBoost"])
```

### Task_3.3: Train Isolation Forest.

```python
from sklearn.ensemble import IsolationForest

# Initialize Isolation Forest
# Contamination is set close to the observed fraud rate (~3.5%)
iso_model = IsolationForest(
    n_estimators=200,
    contamination=0.035,
    random_state=42,
    n_jobs=-1
)

# Isolation Forest is unsupervised, so use the original scaled training features
iso_model.fit(X_train)

# Predict on test set
# Output: 1 = normal, -1 = anomaly
iso_raw_pred = iso_model.predict(X_test)

# Convert to fraud labels: -1 -> 1 (fraud), 1 -> 0 (non-fraud)
iso_pred = (iso_raw_pred == -1).astype(int)

# Decision scores (higher score = more anomalous after negation)
iso_scores = -iso_model.decision_function(X_test)

# Store metrics
results["Isolation Forest"] = {
    "Accuracy": accuracy_score(y_test, iso_pred),
    "Precision": precision_score(y_test, iso_pred, zero_division=0),
    "Recall": recall_score(y_test, iso_pred, zero_division=0),
    "F1-Score": f1_score(y_test, iso_pred, zero_division=0),
    "ROC-AUC": roc_auc_score(y_test, iso_scores),
    "PR-AUC": average_precision_score(y_test, iso_scores)
}

print("Isolation Forest training completed successfully.")
print(results["Isolation Forest"])
```

### Task_3.4: Compare Model Performance.

```python
# Convert results dictionary to DataFrame
results_df = pd.DataFrame(results).T

# Sort by F1-Score (best metric for imbalanced classification)
results_df = results_df.sort_values(
    by="F1-Score",
    ascending=False
)

# Display model comparison table
print("Model Comparison Results:")
display(results_df.round(4))

# Plot comparison chart
plt.figure(figsize=(10, 6))
results_df[["Accuracy", "Precision", "Recall", "F1-Score"]].plot(kind="bar")
plt.title("Model Performance Comparison")
plt.ylabel("Score")
plt.xticks(rotation=0)
plt.legend(loc="best")
plt.tight_layout()

# Save plot
plt.savefig("model_comparison.png", dpi=300)

# Show plot
plt.show()
```

### Task_3.5: Select Best Model and Optimize Classification Threshold.

```python
from sklearn.metrics import precision_recall_curve

# Use LightGBM as the selected best model
# (If another model performs better, update this selection manually.)
best_model_name = "LightGBM"
best_model = lgbm_model
best_proba = lgbm_proba

# Compute precision-recall curve
precision, recall, thresholds = precision_recall_curve(
    y_test,
    best_proba
)

# Calculate F1 score for each threshold
f1_scores = 2 * (precision[:-1] * recall[:-1]) / (
    precision[:-1] + recall[:-1] + 1e-10
)

# Find optimal threshold
best_index = f1_scores.argmax()
optimal_threshold = thresholds[best_index]
best_threshold_f1 = f1_scores[best_index]

print(f"Selected Model: {best_model_name}")
print(f"Optimal Threshold: {optimal_threshold:.4f}")
print(f"Best F1-Score at Optimal Threshold: {best_threshold_f1:.4f}")

# Predictions using optimized threshold
optimized_pred = (best_proba >= optimal_threshold).astype(int)

# Store optimized metrics
optimized_results = {
    "Accuracy": accuracy_score(y_test, optimized_pred),
    "Precision": precision_score(y_test, optimized_pred),
    "Recall": recall_score(y_test, optimized_pred),
    "F1-Score": f1_score(y_test, optimized_pred),
    "ROC-AUC": roc_auc_score(y_test, best_proba),
    "PR-AUC": average_precision_score(y_test, best_proba)
}

print("\nOptimized Model Performance:")
for metric, value in optimized_results.items():
    print(f"{metric}: {value:.4f}")
```

### Task_3.9: Tune the Best Model Using RandomizedSearchCV.

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.utils import resample

# ---------------------------------------------------
# Take a smaller sample from SMOTE training data
# ---------------------------------------------------
# Use up to 100,000 rows to reduce memory usage
sample_size = min(100000, len(X_train_smote))

X_tune, y_tune = resample(
    X_train_smote,
    y_train_smote,
    n_samples=sample_size,
    random_state=42,
    stratify=y_train_smote
)

print("Tuning sample shape:", X_tune.shape)

# ---------------------------------------------------
# Hyperparameter search space
# ---------------------------------------------------
param_distributions = {
    "n_estimators": [100, 200],
    "learning_rate": [0.01, 0.05, 0.1],
    "num_leaves": [31, 63],
    "max_depth": [-1, 6, 10]
}

# ---------------------------------------------------
# Base model
# ---------------------------------------------------
lgbm_tuning_model = LGBMClassifier(
    random_state=42,
    n_jobs=-1
)

# ---------------------------------------------------
# Randomized Search
# ---------------------------------------------------
random_search = RandomizedSearchCV(
    estimator=lgbm_tuning_model,
    param_distributions=param_distributions,
    n_iter=5,          # Reduced for faster execution
    scoring="f1",
    cv=3,
    verbose=1,
    random_state=42,
    n_jobs=1           # Use single process to reduce RAM usage
)

# ---------------------------------------------------
# Fit on sampled data
# ---------------------------------------------------
random_search.fit(X_tune, y_tune)

# ---------------------------------------------------
# Best tuned model
# ---------------------------------------------------
best_tuned_model = random_search.best_estimator_

print("\nBest Parameters:")
print(random_search.best_params_)

print("\nBest Cross-Validation F1-Score:")
print(round(random_search.best_score_, 4))
```

### Task_4.2: Generate Global SHAP Summary Plot (Top 20 Features).

```python
# Use a manageable sample from the test set for SHAP analysis
sample_size = min(2000, len(X_test))
X_shap = X_test.sample(n=sample_size, random_state=42)

print("SHAP sample shape:", X_shap.shape)

# Create SHAP explainer
explainer = shap.TreeExplainer(final_model)

# Compute SHAP values
shap_values = explainer.shap_values(X_shap)

# Handle binary classification output
if isinstance(shap_values, list):
    shap_values_to_plot = shap_values[1]
else:
    # Newer SHAP versions may return a 3D array or Explanation object
    try:
        shap_values_to_plot = shap_values[:, :, 1]
    except Exception:
        shap_values_to_plot = shap_values

# Global SHAP summary plot (top 20 features)
plt.figure()
shap.summary_plot(
    shap_values_to_plot,
    X_shap,
    max_display=20,
    show=False
)

# Save plot
plt.savefig("shap_summary.png", dpi=300, bbox_inches="tight")

# Show plot
plt.show()
```

### Task_4.3: Generate SHAP Waterfall Plots for Three Representative Transactions.

```python
# ---------------------------------------------------
# Select representative transactions:
# 1. Confirmed fraud case
# 2. Borderline case (~0.50 probability)
# 3. Legitimate transaction
# ---------------------------------------------------

# Predicted probabilities on the test set
test_proba = final_model.predict_proba(X_test)[:, 1]

# 1. Confirmed fraud case: highest probability among actual frauds
fraud_indices = y_test[y_test == 1].index
fraud_case_idx = test_proba[
    X_test.index.get_indexer(fraud_indices)
].argmax()
fraud_case = X_test.loc[
    fraud_indices[fraud_case_idx]
]

# 2. Borderline case: probability closest to 0.50
borderline_position = np.argmin(np.abs(test_proba - 0.50))
borderline_case = X_test.iloc[borderline_position]

# 3. Legitimate transaction: lowest probability
legitimate_position = np.argmin(test_proba)
legitimate_case = X_test.iloc[legitimate_position]

# Combine selected cases
cases = {
    "confirmed_fraud": fraud_case,
    "borderline_case": borderline_case,
    "legitimate_transaction": legitimate_case
}

# Generate SHAP waterfall plots
for case_name, case_data in cases.items():
    # Compute SHAP explanation for one row
    explanation = explainer(case_data.to_frame().T)

    # Plot waterfall
    shap.plots.waterfall(
        explanation[0],
        max_display=20,
        show=False
    )

    # Save plot
    plt.savefig(
        f"charts/{case_name}_waterfall.png",
        dpi=300,
        bbox_inches="tight"
    )

    # Show plot
    plt.show()
```

### Task_7.5: Generate Precision-Recall Curve with Optimal Threshold.

```python
# Precision-Recall Curve with Optimal Threshold


import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import precision_recall_curve

# Create predicted probabilities if not already defined
if "y_proba" not in globals():
    y_proba = best_tuned_model.predict_proba(X_test)[:, 1]

# Calculate precision-recall values
precision, recall, thresholds = precision_recall_curve(y_test, y_proba)

# Calculate F1 score for each threshold
f1_scores = 2 * (precision[:-1] * recall[:-1]) / (
    precision[:-1] + recall[:-1] + 1e-10
)

# Find optimal threshold
best_index = np.argmax(f1_scores)
optimal_threshold = thresholds[best_index]
optimal_precision = precision[best_index]
optimal_recall = recall[best_index]

# Create plot
plt.figure(figsize=(8, 6))
plt.plot(recall, precision, linewidth=2, label="Precision-Recall Curve")

# Highlight optimal point
plt.scatter(
    optimal_recall,
    optimal_precision,
    color="red",
    s=100,
    label=f"Optimal Threshold = {optimal_threshold:.4f}"
)

plt.title("Precision-Recall Curve with Optimal Threshold")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.grid(True, alpha=0.3)
plt.legend()

# Save chart
plt.savefig("precision_recall_curve.png", bbox_inches="tight")

# Display chart
plt.show()

# Print key metrics
print(f"Optimal Threshold: {optimal_threshold:.4f}")
print(f"Precision at Optimal Threshold: {optimal_precision:.4f}")
print(f"Recall at Optimal Threshold: {optimal_recall:.4f}")
print(f"Best F1 Score: {f1_scores[best_index]:.4f}")
```

### Task_8.2: Explain Why PR-AUC Matters More Than Accuracy in Fraud Detection.

```python
from IPython.display import Markdown, display

task_8_2 = """
### Why PR-AUC Is More Important Than Accuracy

Fraud detection is a highly imbalanced classification problem, where fraudulent
transactions represent only a small fraction of the total dataset.

#### Limitations of Accuracy
A model that predicts every transaction as legitimate can still achieve very
high accuracy, but it would fail to detect any fraudulent activity.

#### Why PR-AUC Matters
Precision-Recall Area Under the Curve (PR-AUC) focuses specifically on the
performance of the positive class (fraud).

- **Precision** measures how many flagged transactions are truly fraudulent.
- **Recall** measures how many actual fraud cases are detected.
- **PR-AUC** summarizes this trade-off across all thresholds.

#### Business Importance
A high PR-AUC indicates that the model can detect fraud effectively while
minimizing unnecessary investigations and customer disruptions.

#### Conclusion
PR-AUC is a more meaningful metric than accuracy because it directly measures
the model's ability to identify rare fraudulent transactions.
"""

display(Markdown(task_8_2))
```
